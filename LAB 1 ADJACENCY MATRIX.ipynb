```python
#ADJACENCY MATRIX


def addVertex(v):
    global graph
    global vCount
    graph2 = []
    for i in range(0,vCount):
        graph2.append([])
        for j in range(0,vCount):
            graph2[i].append(0)
            
    for i in range(0,vCount-1):
        for j in range(0, vCount-1):
            graph2[i][j] = graph[i][j]
            
    graph = graph2
            
    

def addEdge(s,d,c):
    global vLabels
    source = 0
    for i in range(0, len(vLabels)):
        if s == vLabels[i]:
            source = i
        
        if d == vLabels[i]:
            destination = i
        
    graph[source][destination] = c

def display(graph):
    print(graph)
    


if __name__ == '__main__':
    graph = []
    vCount = 0
    vLabels = []
    option = 0
    end = False
    while(end != True):
        option = int(input("ENTER 1 TO ADD VERTEX, ENTER 2 TO ADD EDGE, ENTER 3 TO DISPLAY GRAPH USING ADJACENCY MATRIX!"))

        if(option == 1):
            v = input("ENTER VERTEX LABEL: ")
            if v not in vLabels:
                vLabels.append(v)
                vCount = vCount + 1
                addVertex(v)
                 
            else:
                print("VERTEX ALREADY PRESENT!")
        elif(option == 2):
            s = input("ENTER SOURCE LABEL: ")
            d = input("ENTER DESTINATION LABEL: ")
            c = int(input("ENTER COST OF EDGE: "))
            
            if s in vLabels and d in vLabels:
                addEdge(s,d,c)
                addEdge(d,s,c)
                
            else:
                print("INVALID VERTEX LABEL GIVEN!")

        elif(option == 3):
            display(graph)
            
        elif(option == 4):
            end == True
            
        else:
            print("INVALID INPUT!")
            
        

    
    
    
```
